```python
#creating an array
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)

zeros_arr = np.zeros((3,3), dtype=int)
print(zeros_arr)

ones_arr = np.ones((3,3))
print(ones_arr)

arange_arr = np.arange(10)
print(arange_arr)

#array manipulation

reshaped_arr = arr.reshape(5,1)
print(reshaped_arr)

sliced_arr = arr[2:4]
print(sliced_arr)

```

```python
#addition of arrays
import numpy as np
arr1 = np.array([1,2,3,4,5])
print(arr1)
arr2 = np.array([5,6,3,7,8])
print(arr2)
print(arr1+arr2)
```

```python
#broadcasting
import numpy as np
arr1 = np.array([1,2,3,4,5])
print(arr1+3)
```

```python
#stack
import numpy as np
arr1 = np.array([1,2,3,4,5])
print(arr1)
arr2 = np.array([5,6,3,7,8])
print(arr2)
c=np.vstack(arr1+arr2)
print(c)
```

```python
import numpy as np
arr1 = np.array([1,2,3,4,5])
print(arr1)
arr2 = np.array([5,6,3,7,8])
print(arr2)
c=np.stack(arr1+arr2)
print(c)
```

```python
#split
arr1 = np.array([1,2,3,4,5,6])
c=np.split(arr1,3)
print(c)
```

```python
#transpose an array
arr1 = np.array([1,2,3,4,5,6,7,8])
arr2=arr1.T
print(arr2)
```

```python
arr1 = np.array([[1,2,3,4],[5,6,7,8]])
arr2=arr1.T
print(arr2)
```

```python
#linear algebra with numpy
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
c=np.dot(arr1,arr2)
print(c)
```

```python
#calculating eigen values and eigen vector
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
c=np.dot(arr1,arr2)
print(c)
d=np.linalg.eig(c)
print(d)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_temperature = np.array([29, 29, 29, 28, 28, 29, 29, 29, 29, 30, 30, 31, 31, 32, 32, 31, 30, 31, 30, 31, 32, 33, 33, 32, 31, 30, 31, 30, 31, 31, 31])
dates = list(range(1,3))
threshold = np.mean(data_temperature)
print(threshold)

min_temp_index = np.argmin(data_temperature)
max_temp_index = np.argmax(data_temperature)
min_temp_date = dates[min_temp_index]
max_temp_date = dates[max_temp_index]

plt.figure(figsize=(12, 6))

plt.plot(dates, data_temperature, color='purple', linewidth=2, label='January', marker='.', ms=15, mec='b')
plt.title('Daily Temperature Graph')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.legend(loc='best')

threshold_line = [threshold] * 31
plt.plot(dates, threshold_line, color='red', linestyle='dotted', linewidth=2, label='Threshold'

plt.scatter([min_temp_date, max_temp_date], [np.min(data_temperature), np.max(data_temperature)], color='green', marker='o', label='Min/Max Temperature')

plt.text(min_temp_date, np.min(data_temperature), f'Min: ({min_temp_date}, {np.min(data_temperature)})', color='green', ha='right', va='bottom')
plt.text(max_temp_date, np.max(data_temperature), f'Max: ({max_temp_date}, {np.max(data_temperature)})', color='green', ha='left', va='top')

plt.title('Temperature Analysis')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Display the first few rows of the dataset to understand its structure
print(anscombe.head())

# Plot the four datasets in the anscombe dataset
sns.set(style="ticks")
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=anscombe,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
plt.show()
```
